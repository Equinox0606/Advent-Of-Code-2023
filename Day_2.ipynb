### Day 2: Cube Conundrum
[Day 2]((https://adventofcode.com/2023/day/2))

You're launched high into the atmosphere! The apex of your trajectory just barely reaches the surface of a large island floating in the sky. You gently land in a fluffy pile of leaves. It's quite cold, but you don't see much snow. An Elf runs over to greet you.

The Elf explains that you've arrived at Snow Island and apologizes for the lack of snow. He'll be happy to explain the situation, but it's a bit of a walk, so you have some time. They don't get many visitors up here; would you like to play a game in the meantime?

As you walk, the Elf shows you a small bag and some cubes which are either red, green, or blue. Each time you play this game, he will hide a secret number of cubes of each color in the bag, and your goal is to figure out information about the number of cubes.

To get information, once a bag has been loaded with cubes, the Elf will reach into the bag, grab a handful of random cubes, show them to you, and then put them back in the bag. He'll do this a few times per game.

You play several games and record the information from each game (your puzzle input). Each game is listed with its ID number (like the 11 in Game 11: ...) followed by a semicolon-separated list of subsets of cubes that were revealed from the bag (like 3 red, 5 green, 4 blue).

For example, the record of a few games might look like this:
```
Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green
Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue
Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red
Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red
Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green
```

In game 1, three sets of cubes are revealed from the bag (and then put back again). The first set is 3 blue cubes and 4 red cubes; the second set is 1 red cube, 2 green cubes, and 6 blue cubes; the third set is only 2 green cubes.

The Elf would first like to know which games would have been possible if the bag contained only 12 red cubes, 13 green cubes, and 14 blue cubes?

In the example above, games 1, 2, and 5 would have been possible if the bag had been loaded with that configuration. However, game 3 would have been impossible because at one point the Elf showed you 20 red cubes at once; similarly, game 4 would also have been impossible because the Elf showed you 15 blue cubes at once. If you add up the IDs of the games that would have been possible, you get 8.

Determine which games would have been possible if the bag had been loaded with only 12 red cubes, 13 green cubes, and 14 blue cubes. What is the sum of the IDs of those games?

### Thought process

 1. **Class Initialization (`__init__`)**:
   - The constructor initializes the class with an `input_file` path and an empty dictionary `games` to store the maximum count of each color for each game.

 2. **Reading Input (`read_input`)**:
   - This method reads the input file line by line.
   - Each line (representing a game) is parsed using the `parse_game_line` method.
   - The parsed data, which includes the game ID and the maximum counts of each color for that game, is stored in the `games` dictionary.

 3. **Parsing Game Line (`parse_game_line`)**:
   - It splits each line into a game ID and color-count information.
   - The game ID is extracted using a generator expression that filters out the numeric part from the game ID string.
   - The color-count pairs are processed, and the maximum count for each color is determined. This step is crucial to understand if a game can be played with the given number of cubes.

 4. **Checking Game Possibility (`is_game_possible`)**:
   - This method determines if a game is possible given the maximum counts of each color shown and the specified number of cubes (12 red, 13 green, 14 blue).
   - It uses the `all()` function to ensure that for each color, the maximum number shown in any subset does not exceed the available number of cubes. If it does, the game is not possible.

 5. **Calculating Sum of Possible Game IDs (`calculate_possible_games_sum`)**:
   - This method iterates through each game stored in `games`.
   - It checks if each game is possible using the `is_game_possible` method.
   - If a game is possible, its ID is added to the sum.

 6. **Solving the Puzzle (`solve_puzzle`)**:
   - This method combines all the functionalities: it reads the input, calculates the sum of IDs of possible games using the predetermined number of cubes, and returns the sum.

 7. **Running the Code**:
   - An instance of `CubeConundrum` is created with the path to the input file.
   - The `solve_puzzle` method is called to get the sum of IDs of all possible games, which is then printed.
    
    
### Summary of the Thought Process:

1. **Input File Processing**: The class is specifically designed to interpret and process a text file, where each line represents a game with unique ID and color-cube counts.

2. **Game Data Parsing**: The essential function is to parse each line to extract the game ID and determine the maximum counts of red, green, and blue cubes shown in that game.

3. **Game Possibility Evaluation**: The core objective is to evaluate which games are feasible under a given constraint (specific numbers of red, green, and blue cubes).

4. **Structured Game Analysis**: The class provides a methodical approach to reading the input file, analyzing each game's data, and calculating the sum of IDs of the possible games.

````python
class CubeConundrum:
    def __init__(self, input_file):
        self.input_file = input_file
        self.games = {}

    def read_input(self):
        """
        Reads the input file and stores the information of each game.
        """
        with open(self.input_file, 'r') as file:
            for line in file:
                game_id, colour_counts = self.parse_game_line(line)
                self.games[game_id] = colour_counts
    
    def parse_game_line(self, line):
        """
        Parses a single line of input and returns the game ID and the subsets of cubes shown.
        """
        # Split the line into game ID and colour-count pairs
        game_id, colours_info = line.split(':')
        game_id = int(next(s for s in game_id.split() if s.isdigit()))

        # Process the colour-count pairs
        colour_counts = {}
        pairs = colours_info.split(';')
        for pair in pairs:
            # Process each colour-count pair
            colour_count_pairs = pair.strip().split(',')
            for colour_count in colour_count_pairs:
                count, colour = colour_count.strip().split()
                count = int(count)

                # Update the maximum count for each colour
                colour_counts[colour] = max(colour_counts.get(colour, 0), count)

        return game_id, colour_counts

    def is_game_possible(self, colour_counts, red_cubes, green_cubes, blue_cubes):
        """
        Determines if a game is possible with the given number of red, green, and blue cubes.
        """
        return all([
                colour_counts.get('red', 0) <= red_cubes,
                colour_counts.get('green', 0) <= green_cubes,
                colour_counts.get('blue', 0) <= blue_cubes
            ])

    def calculate_possible_games_sum(self, red_cubes, green_cubes, blue_cubes):
        """
        Calculates the sum of the IDs of the games that are possible with the given cube configuration.
        """
        possible_games_sum = 0
        for game_id, counts in self.games.items():
            if self.is_game_possible(counts, red_cubes, green_cubes, blue_cubes):
                possible_games_sum += game_id
        return possible_games_sum

    def solve_puzzle(self):
        """
        Solves the puzzle by reading the input and determining the sum of IDs of possible games.
        """
        self.read_input()
        red_cubes = 12
        green_cubes = 13
        blue_cubes = 14
        return self.calculate_possible_games_sum(red_cubes, green_cubes, blue_cubes)

input_file = r'path_to_file.txt'
conundrum = CubeConundrum(input_file)
result = conundrum.solve_puzzle()
print("Sum of possible game IDs:", result)
````

### Part Two
The Elf says they've stopped producing snow because they aren't getting any water! He isn't sure why the water stopped; however, he can show you how to get to the water source to check it out for yourself. It's just up ahead!

As you continue your walk, the Elf poses a second question: in each game you played, what is the fewest number of cubes of each color that could have been in the bag to make the game possible?

Again consider the example games from earlier:
```
Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green
Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue
Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red
Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red
Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green
In game 1, the game could have been played with as few as 4 red, 2 green, and 6 blue cubes. If any color had even one fewer cube, the game would have been impossible.
Game 2 could have been played with a minimum of 1 red, 3 green, and 4 blue cubes.
Game 3 must have been played with at least 20 red, 13 green, and 6 blue cubes.
Game 4 required at least 14 red, 3 green, and 15 blue cubes.
Game 5 needed no fewer than 6 red, 3 green, and 2 blue cubes in the bag.
```
The power of a set of cubes is equal to the numbers of red, green, and blue cubes multiplied together. The power of the minimum set of cubes in game 1 is 48. In games 2-5 it was 12, 1560, 630, and 36, respectively. Adding up these five powers produces the sum 2286.

For each game, find the minimum set of cubes that must have been present. What is the sum of the power of these sets?

### Thought process:

1. **Initialization (`__init__`)**:
   - The class initializes with the path to an input file and prepares a dictionary, `games`, to store game data.

2. **Reading and Parsing Input (`read_input` and `parse_game_line`)**:
   - `read_input` reads the file line by line.
   - `parse_game_line` splits each line into game ID and color-cube counts, extracting the maximum counts of each color in a game and storing this data with the game ID.

3. **Calculating Power Set (`calculate_power_set`)**:
   - This method calculates the product of the maximum counts of red, green, and blue cubes for each game.
   - It interprets these counts as dimensions of a 'power set', reflecting the minimum number of cubes needed in the game.

4. **Summing Game Values (`calculate_games_sum`)**:
   - Instead of checking game feasibility, this method sums the 'power set' values from `calculate_power_set` for all games.
   - This sum represents the cumulative minimum number of cubes across all games.

5. **Solving the Puzzle (`solve_puzzle`)**:
   - Combines the steps: reads input, calculates the total sum of the 'power set' values, and returns this sum.

6. **Execution**:
   - An instance of the class is created with the input file, and the puzzle is solved to get the sum of all 'power set' values, which is then printed.

### Summary of Thought Process:

1. **Input File Handling**: The class is designed to read and interpret a text file, where each line represents a game's data with ID and color-cube counts.

2. **Data Parsing and Processing**: The key function is to parse each line to obtain game IDs and calculate the maximum counts of each cube color.

3. **Power Set Calculation**: Computing a 'power set' value—which is the bare minimum of cubes needed for every game—is a new feature.

4. **Cumulative Summation**: The primary goal shifts to calculating the cumulative sum of these 'power set' values across all games.

5. **Structured Approach**: The implementation follows a systematic method to read, process, and sum the data, providing a comprehensive solution to the puzzle.

````python
class CubeConundrum:
    def __init__(self, input_file):
        self.input_file = input_file
        self.games = {}

    def read_input(self):
        """
        Reads the input file and stores the information of each game.
        """
        with open(self.input_file, 'r') as file:
            for line in file:
                game_id, colour_counts = self.parse_game_line(line)
                self.games[game_id] = colour_counts
    
    def parse_game_line(self, line):
        """
        Parses a single line of input and returns the game ID and the subsets of cubes shown.
        """
        # Split the line into game ID and colour-count pairs
        game_id, colours_info = line.split(':')
        game_id = int(next(s for s in game_id.split() if s.isdigit()))

        # Process the colour-count pairs
        colour_counts = {}
        pairs = colours_info.split(';')
        for pair in pairs:
            # Process each colour-count pair
            colour_count_pairs = pair.strip().split(',')
            for colour_count in colour_count_pairs:
                count, colour = colour_count.strip().split()
                count = int(count)

                # Update the maximum count for each colour
                colour_counts[colour] = max(colour_counts.get(colour, 0), count)

        return game_id, colour_counts
    
    def calculate_power_set(self, colour_counts):
        """
        Calculates the power set of minimun number of cubes that must be present
        """
        return colour_counts.get('red') * colour_counts.get('green') * colour_counts.get('blue')

    def calculate_games_sum(self, red_cubes, green_cubes, blue_cubes):
        """
        Calculates the sum of the IDs of the games that are possible with the given cube configuration.
        """
        games_sum = 0
        for game_id, counts in self.games.items():
            games_sum += self.calculate_power_set(counts)
        return games_sum

    def solve_puzzle(self):
        """
        Solves the puzzle by reading the input and determining the sum of IDs of possible games.
        """
        self.read_input()
        red_cubes = 12
        green_cubes = 13
        blue_cubes = 14
        return self.calculate_games_sum(red_cubes, green_cubes, blue_cubes)

input_file = r'path_to_file.txt'
conundrum = CubeConundrum(input_file)
result = conundrum.solve_puzzle()
print("Sum of all power sets:", result)
````
